```python
 # Escrever um programa que o usuário digita o nome e idade:

 # Identificar se o usuário é criança, adolescente ou adulto.

 # > 12 criança
 # > 18 adolescente
 # < adulto

 # Saída (você é...)

 # Entrada

nome = str (input ("digite seu nome"))
idade = int (input ("digite sua idade"))

 # Processamento / Saída

if idade <= 12:
  print (f"ola {nome}, voce e criança")

elif idade <= 18:
  print (f"ola {nome}, voce e adolescente")

else:
  print (f"ola {nome}, voce e adulto")
```
