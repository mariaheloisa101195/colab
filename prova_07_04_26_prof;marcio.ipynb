```python
 # Escrever um programa que o usuário digita o nome e idade:

  #Identificar se o usuário é criança, adolescente ou adulto.

  #> 12 criança
  #> 18 adolescente
  #< adulto

   # Saída (você é...)

  # Entrada

nome = str  (input ("digite seu nome"))
idade = int (input ("digite sua idade"))

# Processamento / Saída


if idade <= 17:
  print (f"ola {nome}, voce é menor de idade")

elif idade <= 59:
  print (f"ola {nome}, voce é adulto")

else:
  print (f"ola {nome}, voce é idoso")
```

```python
 # Operadores aritimeticos

 # pedir ao usuario para digitar 1 número

  #Identificar se o usuario tem direito a desconto.

  #> 100 sem desconto
  #> 500 com desconto
  #< 500 20% de desconto

 # Entrada



valor = float(input("Digite o valor da compra: "))

if valor <= 100:
    desconto = 0
elif valor <= 500:
    desconto = valor * 0.10
else:
    desconto = valor * 0.20

valor_final = valor - desconto

print("Valor final:", valor_final)
```

```python
 # Defina a média escolar de um aluno


# Leitura da média
media = float(input("Digite a média final: "))

# Classificação
if media >= 7:
    print("Aprovado")
elif media >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

```python
 #Cálculo de salário com bônus


 # Leitura dos dados

salario = float(input("Digite o salário: "))
pontuacao = int(input("Digite a pontuação de desempenho (0 a 10): "))

# Cálculo do bônus

if pontuacao < 5:
    bonus = 0
elif pontuacao <= 7:
    bonus = salario * 0.05
elif pontuacao <= 9:
    bonus = salario * 0.10
else: # pontuação == 10
    bonus = salario * 0.15

# Resultado
salario_final = salario + bonus

print(f"Bônus: R$ {bonus:.2f}")
print(f"Salário final: R$ {salario_final:.2f}")
```
