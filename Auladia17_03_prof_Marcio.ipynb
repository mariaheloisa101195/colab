```python
# Iperadores Aritiméticos

# pedir ao usuario para digitar 2 números

 # Entrada

a = int (input("digite um número: "))
b = int (input("digite um número: "))

 # Processamento

 # Soma

Soma = a + b
Subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b
 # Saída

print("")
print("---------------- OPERADORES ARITIMÉTICOS-----------------")
print("")
print("-----------------------SOMA----------------------------")
print("")
print(f"O resultado da soma de {a} mais {b} é igual a : {Soma} ")
print("")
print("-------------------------------------------------------")

print("-----------------------SUBTRAÇÃO----------------------------")
print("")
print(f"O resultado da DIFERENÇA de {a} menos {b} é igual a : {Subtracao} ")
print("")
print("-------------------------------------------------------")

print("-----------------------MULTIPLICAÇÃO----------------------------")
print("")
print(f"O resultado da MULTIPLICAÇÃO de {a} vezes {b} é igual a : {multiplicacao} ")
print("")
print("-------------------------------------------------------")

print("-----------------------DIVISÃO----------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} DIVIDIDO por {b} é igual a : {divisao} ")
print("")
print("-------------------------------------------------------")

print("-----------------------DIVISÃO----------------------------")
print("")
print(f"O resultado da DIVISÃOc INTEIRA de {a} DIVIDINDO POR {b} é igual a : {div_int} ")
print("")
print("-------------------------------------------------------")

print("-----------------------Resto----------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} DIVIDINDO {b} o resto é igual a : {resto} ")
print("")
print("-------------------------------------------------------")

print("-----------------------POTENCIAÇÃO----------------------------")
print("")
print(f"O resultado da POTENCIAÇÃO de {a} ELEVADO {b} é igual a : {potencia} ")
print("")
print("-------------------------------------------------------")
```
